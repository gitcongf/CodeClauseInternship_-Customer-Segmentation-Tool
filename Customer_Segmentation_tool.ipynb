```python
X= df.iloc[:,[3,4]].values #annual income col & spending score col
```

```python
print(X)
```

```python
wcss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('Elbow point Graph')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmean = KMeans(n_clusters=10, init='k-means++', random_state=0)
Y= kmeans.fit_predict(X)
print(Y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1'  )
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2'  )
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3'  )
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4'  )
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5'  )
plt.scatter(X[Y==5,0], X[Y==5,1], s=50, c='black', label='Cluster 6'  )
plt.scatter(X[Y==6,0], X[Y==6,1], s=50, c='orange', label='Cluster 7'  )
plt.scatter(X[Y==7,0], X[Y==7,1], s=50, c='cyan', label='Cluster 8'  )
plt.scatter(X[Y==8,0], X[Y==8,1], s=50, c='purple', label='Cluster 9'  )
plt.scatter(X[Y==9,0], X[Y==9,1], s=50, c='magenta', label='Cluster 10'  )
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=150, c='brown', label='Centroids')
plt.title('Cust Grps')
plt.xlabel('Yearly inc')
plt.ylabel('Spend Score')
plt.show()
```
